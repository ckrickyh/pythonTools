```python
from pathlib import Path

pathDir = input('Provide a file directory')
# Specify the directory containing the files
directory_path = Path(pathDir)

# Dictionary of abbreviations to replace
abbreviations = {
    '_cb': 'cross branch',
    '_sp': 'sparse',
    '_db': 'dead branch',
    '_WholeView': '',
    ' (2)': '',
    #'__wh': '_WholeView'
    #' (1)': '',
}

# Function to replace abbreviations in a filename
def replace_abbreviations(filePath):
    filename = Path(filePath).stem 

    for abbr, full in abbreviations.items():
        
        filename = filename.replace(abbr, full)
 
    # Append ' (1)' if '(' is not in the filename
    #if '(' not in filename:
     #      filename+= ' (1)'
    
    return filename  # Append the original suffix
    

for fl in directory_path.iterdir():
    if fl.is_file():
        
        new_name = replace_abbreviations(fl) + fl.suffix
        
        # Rename the file
        fl.rename(directory_path / new_name)
        print(f'Renamed: {fl.name} to {new_name}')
        print('done')
```
